```python
# Importing packages

import pandas as pd
import requests
from bs4 import BeautifulSoup

pd.options.mode.chained_assignment = None
```

```python
# Defining functions to scrape from VGChartz

def get_soup(url):
    '''
    Parse url with BeautifulSoup
    
    Inputs
    ----
    url (string)
        url of page to parse
    
    Outputs
    ----
    soup (BeautifulSoup)
        Soup object
    '''

    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    return soup

def scrape_page(page): 
    '''
    Scrapes the page from VGChartz

    Inputs
    ----
    page (int)
        page number

    Outputs
    ----
    dfs (list)
        list of dataframes with page details
    '''
    print(page)
    
    #parse url and scrape name and developer
    url = f"https://www.vgchartz.com/games/games.php?page={page}name=&keyword=&console=&region=All&developer=&publisher=&goty_year=&genre=&boxart=Both&banner=Both&ownership=Both&showmultiplat=No&results=200&order=Sales&showtotalsales=0&showtotalsales=1&showpublisher=0&showvgchartzscore=0&shownasales=0&shownasales=1&showdeveloper=0&showdeveloper=1&showcriticscore=0&showpalsales=0&showpalsales=1&showreleasedate=0&showreleasedate=1&showuserscore=0&showjapansales=0&showjapansales=1&showlastupdate=0&showothersales=0&showothersales=1&showshipped=0"
    soup = get_soup(url)
    sales_dict = {}
    sales_dict['name'] = list(map(lambda x: x.get_text().strip(),soup.find('div',{'id':'generalBody'}).select('tr > td:nth-of-type(3) > a:nth-of-type(1)')))
    sales_dict['developer'] = list(map(lambda x: x.get_text().strip(),soup.find('div',{'id':'generalBody'}).select('tr > td:nth-of-type(5)')))
    sales_dict['date'] = list(map(lambda x: x.get_text().strip(),soup.find('div',{'id':'generalBody'}).select('tr > td:nth-of-type(11)')))
    sales_dict['platform'] = list(map(lambda x: x.get_attribute_list('alt')[0],soup.find('div',{'id':'generalBody'}).select('tr > td:nth-of-type(4) > img')))

    #scrape sales
    for i,k in zip(range(6,11),['total','NA','PAL','japan','other']):
        sales_dict[k+'_sales'] = list(map(lambda x: x.get_text(),soup.find('div',{'id':'generalBody'}).select(f'tr > td:nth-of-type({i})')))
    #pd.DataFrame(sales_dict).to_csv(f'./tempdata/page{page}.csv',index=False)
    return(pd.DataFrame(sales_dict))

def main():
    '''
    Scrape each page and save
    '''
    scraped = []
    for i in range(1,314):
        scraped.append(scrape_page(i))
    
    pd.concat(scraped,ignore_index=True).to_csv('gamesales_150223.csv',index=False)


```

```python
main()
```
